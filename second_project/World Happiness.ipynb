```python
# Removing the skewness by using Cube root square.
df['Region'] = np.cbrt(df['Region'])

# Removing the skewness by using Cube root square.
df['Standard Error'] = np.cbrt(df['Standard Error'])

# Removing the skewness by using Cube root square.
df['Economy (GDP per Capita)'] = np.cbrt(df['Economy (GDP per Capita)'])

# Removing the skewness by using Cube root square.
df['Family'] = np.cbrt(df['Family'])

# Removing the skewness by using Cube root square.
df['Health (Life Expectancy)'] = np.cbrt(df['Health (Life Expectancy)'])

# Removing the skewness by using Cube root square.
df['Freedom'] = np.cbrt(df['Freedom'])

# Removing the skewness by using Cube root square.
df['Trust (Government Corruption)'] = np.cbrt(df['Trust (Government Corruption)'])

# Removing the skewness by using Cube root square.
df['Generosity'] = np.cbrt(df['Generosity'])

# Removing the skewness by using Cube root square.
df['Dystopia Residual'] = np.cbrt(df['Dystopia Residual'])
```

```python
# Removing outliers
from scipy import stats

for col in df.columns:
    z_scores = stats.zscore(df[col])
    abs_z_scores = np.abs(z_scores)
    filtered_entries = (abs_z_scores < 3)
    df = df[filtered_entries]
```

```python
# Separating feature and label
x = df.drop('Happiness Score', axis=1)
y = df['Happiness Score']
```

```python
# Feature Scaling using Standard Scalarization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
# Checking Variance Inflation Factor(VIF)
# Checking VIF values to confirm whether multicollinearity exists or not.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['VIF Values'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
vif['Feature Values'] = x.columns
```

```python
# So, we can clearly see that there is multicollinearity present in Happiness Rank, so we need to drop the column.
vif
```

```python
# As we have observed a high correlation that affects the target value, dropping that column can resolve the issue.  
df.drop('Happiness Rank', axis=1, inplace=True)
```

```python
# Now again checking high correlation that was affecting the target value.
# Separating feature and label
x = df.drop('Happiness Score', axis=1)
y = df['Happiness Score']

# Feature Scaling using Standard Scalarization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)

# Checking Variance Inflation Factor(VIF)
# Checking VIF values to confirm whether multicollinearity exists or not.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['VIF Values'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]
vif['Feature Values'] = x.columns

vif
# So, we can clearly see that there is no multicollinearity present, so we don't need to drop any columns.
```

### Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=.2, shuffle=True)
```

# RandomForestRegressor

```python
RFR=RandomForestRegressor()
RFR.fit(x_train, y_train)
```

```python
y_pred=RFR.predict(x_test)
mean_squared_error(y_test,y_pred)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

# LinearRegression

```python
LR=LinearRegression()
LR.fit(x_train, y_train)
y_pred=LR.predict(x_train)
mean_squared_error(y_train,y_pred)
```

```python
y_pred=LR.predict(x_test)
mean_squared_error(y_test,y_pred)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

# SVR

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train, y_train)
y_pred=svr.predict(x_train)
mean_squared_error(y_train,y_pred)
```

```python
y_pred=svr.predict(x_test)
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

# KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
KNR=KNeighborsRegressor()
KNR.fit(x_train, y_train)
y_pred=KNR.predict(x_train)
mean_squared_error(y_train,y_pred)
```

```python
y_pred=KNR.predict(x_test)
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

#### LinearRegression seems to be the best model for your data as it achieves the highest 𝑅2 score, indicating the best fit to your data among the models compared.
